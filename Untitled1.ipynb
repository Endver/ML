```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate sample data (replace with your own dataset)
X = np.random.rand(100, 2) * 10  # 100 points in 2D space

# Standardize the dataset (DBSCAN is sensitive to scale)
X_scaled = StandardScaler().fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps and min_samples as needed
labels = dbscan.fit_predict(X_scaled)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.xlabel("average_track_popularity")
plt.ylabel("top_genre")
plt.title("DBSCAN Clustering")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score (only for non-noise points)
if len(set(labels)) > 1 and -1 in labels:  # Ignore noise points for scoring
    score = silhouette_score(X_scaled, labels)
    print(f"Silhouette Score: {score}")
```
